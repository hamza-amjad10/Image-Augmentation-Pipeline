```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('/content/images/cat.jpeg',target_size=(256,256))
```

```python
img
```

```python
!mkdir augmentation_images
```

```python
new_data = ImageDataGenerator(
    rotation_range=30,      # rotate the image randomly between -30 and +30 degrees
    shear_range=0.2,        # apply shear transformation (slant the image left/right)
    zoom_range=0.2,         # zoom in or out randomly (20%)
    horizontal_flip=True,   # flip the image horizontally (mirror)
    width_shift_range=0.2,  # shift the image left or right by up to 20% of the width
    height_shift_range=0.2, # shift the image up or down by up to 20% of the height
    fill_mode='nearest'     # fill empty pixels after transformation using nearest pixel values
)
```

```python
img=image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,256,256,3)
```

```python
i=0
for output in new_data.flow(input_batch,batch_size=1,save_to_dir="augmentation_images"):
  i=i+1

  if i==10:
    break
```
